```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
 final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
plt.pie(country_value[:3], labels=country_names[:3], autopct='%1.2f%%')
```

- Observation: Zomato Maximum records or Transcation are from India
After that USA and then United Kingdoms

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

### Observation:
1. When Rating are between 4.9 to 4.5 ----> Excellent
2. When Rating are between 4.4 to 4.0 ----> Very Good
3. When Rating are between 3.9 to 3.5 ----> Good
4. When Rating are between 3.4 to 3.0 ----> Average
5. When Rating are between 2.9 to 2.5 ----> Below Average
6. When Rating are between 2.5 to 2.0 ----> Poor

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(
    x='Aggregate rating', 
    y='Rating count', 
    data=ratings, 
    hue='Rating color',
    palette=['white','red','orange','yellow','green','green'],
    legend=False
)
```

### Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 3.4 to 2.5

```python
##Count Plot
sns.countplot(x='Rating color',data=ratings, palette=['white','red','orange','yellow','green','green'])
```
