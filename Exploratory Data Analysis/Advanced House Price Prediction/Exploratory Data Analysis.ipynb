### 1. Missing Values

```python
#list of features which has missing values.
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4), '%missing values')
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

```python
for feature in features_with_na:
    data = dataset.copy()
    
#Make a  variable that indicates if observation is missing then value is 1 or otherwise 0.
    data[feature] = np.where(data[feature].isnull(), 1, 0 )

#calculate the mean Sale Price where information is missing or not

    data.groupby(feature)['SalePrice'].median().plot.bar(color=['blue','orange'])
    plt.title(feature)
    plt.show()
```

### Numerical values:-

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
# 'O' stands for Object dtype (usually strings / categorical data)

print('Number of numerical variables: ', len(numerical_features))

dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
#Explore the content of these year variables.

for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Relation between year the house is sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs YearSold')
```

```python
year_feature
```

```python
# Compare between All years feature and SalesPrice

for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()

        data[feature] = data['YrSold']-data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

#### Numerical variables are usually 2 types:-
- 1. Continous variable
- 2. Discrete variable

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature + ['Id']]
print("Discrete Variavle Count : {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
#Relationship between discrete_feature and SalePrice
for feature in discrete_feature:
    data = dataset.copy()
    grouped = data.groupby(feature)['SalePrice'].median()
    colors = plt.cm.tab20(np.linspace(0, 1, len(grouped)))
    grouped.plot.bar(color=colors)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### Continuous Variable

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature + ['Id']]
print("Continuoues Feature Count: {}".format(len(continuous_feature)))
```

```python
# Analysis the continuous value by creating histograms to understand the distribution

for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins = 25, color='steelblue', edgecolor='black'      )
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

### Logarithmic Transformation:-

```python
for feature in continuous_feature:
    data= dataset.copy()
    if (data[feature] == 0).any():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers:-


```python
for feature in continuous_feature:
    data = dataset.copy()
    if (data[feature] == 0).any():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
