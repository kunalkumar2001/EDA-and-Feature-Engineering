#### One HotEncoding with many categories in a feature

```python
df=pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df.head()
```

```python
df['X0'].value_counts()
```

```python
df['X0'].unique()
```

```python
len(df['X0'].unique())
```

```python
## All columns length value
for i in df.columns:
    print(len(df[i].unique()))
```

```python
df.X1.value_counts()
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
lst_10=list(lst_10)
```

```python
lst_10
```

```python
for categories in lst_10:
    df[categories] = np.where(df['X1']==categories,1,0)
```

```python
lst_10.append('X1')
```

```python
df[lst_10]
```
